## Ответьте на вопросы:

- В чем принципиальное отличие гибридных рекомендательных систем от коллаборативной филтьтрации?

Коллаборативной филтьтрация предполагает, что у пользователей со схожими вкусами в прошлом будут похожие вкусы и в будущем. Например Netflix использует вариант CF, чтобы рекомендовать фильмы и шоу, которые могут вам понравиться, на основе аналогичных вкусов пользователей в аналогичных фильмах. Контентная фильтрация предполагает, что в будущем пользователю понравятся элементы, которые имеют общие черты, такие как бренд, состав, жанр и т.д., с элементами, которые ему нравились в прошлом.
