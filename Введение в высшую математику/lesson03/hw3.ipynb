### 4. Задание (на листочке)
1) Пусть задана плоскость: A1x + B1y + C1z + D1 = 0

Напишите уравнение плоскости, параллельной данной и проходящей через начало координат.

2)  Пусть задана плоскость: A1x + B1y + C1z + D1 = 0
и прямая: 
(x - x1)/(x2 - x1) = (y - y1)/(y2 - y1) = (z - z1)/(z2 - z1)

Как узнать, принадлежит прямая плоскости или нет?

### Решение:
1) Так как плоскость проходит через начало координат, то 0+0+0+D2 = 0, следовательно D2 = 0. Так как для параллельности плоскости необходимо, чтобы все углы сохранялись, то уравнение параллельной плоскости будет: A1кx + B1кy + C1кz = 0 или A1x + B1y + C1z = 0

2) Необходимо выполнить проверку: A1x1 + B1y1 + C1z1 + D1 = A1x2 + B1y2 + C1z2 + D1 = 0. Если тождество соблюдается, тогда прямая принадлежит плоскости и также плоскости принадлежат обе точки.

1) Нарисуйте трехмерный график двух параллельных плоскостей.

```
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X,Y)
Z = X + Y
ax.plot_wireframe(X,Y,Z)
ax.plot_wireframe(X,Y,Z+10)
```
