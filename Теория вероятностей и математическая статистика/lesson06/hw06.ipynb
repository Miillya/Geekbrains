#### 1.Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks):  
**zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110],  
ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832].  
Найдите ковариацию этих двух величин с помощью элементарных действий, а затем с помощью функции cov из numpy. Полученные значения должны быть равны.  
Найдите коэффициент корреляции Пирсона с помощью ковариации и среднеквадратичных отклонений двух признаков, а затем с использованием функций из библиотек numpy и pandas.**

```python
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array ([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])

cov_m = (zp*ks).mean() - zp.mean()*ks.mean()
cov_np = np.cov(zp, ks, ddof=0)
cor_m = cov_m / (np.std(zp)*np.std(ks))
cor_np = np.corrcoef(zp, ks)

print(f'Ковариация с помощью элементарных действий: {cov_m}')
print(f'Ковариация с помощью функции cov из numpy: {cov_np}')
print(f'Коэффициент корреляции Пирсона с помощью ковариации и среднеквадратичных отклонений двух признаков: {cor_m}')
print(f'Коэффициент корреляции Пирсона с использованием функций из библиотек numpy и pandas: {cor_np}')
```
